```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score
```

```python
Target=df['Y']
Features=df.drop('Y',axis=1)
```

```python
le=LabelEncoder()
Target=le.fit_transform(Target)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(Features,Target,test_size=0.2,random_state=100)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Building ANN Model

```python
ann_model=Sequential()
ann_model.add(Dense(16,activation='relu',input_shape=(60,)))
ann_model.add(Dense(1,activation='sigmoid'))
ann_model.summary()
```

 ## Compile the model

```python
ann_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
ann_model.summary()
```

## Training the model

```python
ann_model.fit(X_train,Y_train,epochs=100,batch_size=16)
```

```python
Y_pred_prob=ann_model.predict(X_test)
y_pred = (Y_pred_prob > 0.5).astype(int)
acc = accuracy_score(Y_test, y_pred)
prec = precision_score(Y_test, y_pred)
rec = recall_score(Y_test, y_pred)
f1 = f1_score(Y_test, y_pred)

print("Accuracy:", acc)
print("Precision:", prec)
print("Recall:", rec)
print("F1-Score:", f1)
```

## HYPERPARAMETER SEARCH

```python
from tensorflow.keras.optimizers import Adam
results=[]
best_score=0
best_params=None

neurons_list=[8,16,32]
layer_list=[1,2,3]
activation_list=['relu','tanh']
learning_rates = [0.001, 0.01]
batch_sizes = [16, 32]

for neurons in neurons_list:
    for layers in layer_list:
        for activation in activation_list:
            for lr in learning_rates:
                for batch in batch_sizes:
                    print(f"\nTesting: {neurons} neurons | {layers} layers | {activation} | lr={lr} | batch={batch}")

                    ann_model=Sequential()
                    ann_model.add(Dense(64,activation=activation,input_shape=(60,)))
                    for i in range(layers - 1):
                        ann_model.add(Dense(neurons, activation=activation))

                    ann_model.add(Dense(1,activation='sigmoid'))

                    ann_model.compile(optimizer=Adam(learning_rate=lr),loss='binary_crossentropy',metrics=['accuracy'])

                    ann_model.fit(X_train,Y_train,epochs=50,batch_size=batch)

                    y_pred_prob=ann_model.predict(X_test)
                    y_pred=(y_pred_prob>0.5).astype(int)
                    acc = accuracy_score(Y_test, y_pred)
                    prec = precision_score(Y_test, y_pred)
                    rec = recall_score(Y_test, y_pred)
                    f1 = f1_score(Y_test, y_pred)
                    results.append((neurons, layers, activation, lr, batch, acc))

                    if acc > best_score:
                        best_score=acc
                        best_params=(neurons, layers, activation, lr, batch)
                        print("Accuracy:", acc)

print("\nBEST ACCURACY:", best_score)
print("BEST PARAMETERS:", best_params)
```

## In this experiment, I performed hyperparameter tuning to improve the performance of the ANN model for Sonar classification.
